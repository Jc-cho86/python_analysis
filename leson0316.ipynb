```python
import pandas as pd
import numpy as np
# 建立一個包含 NaN 的 DataFrame
data = {'col1': [1, 2, np.nan, 4], 
        'col2': ['a', np.nan, 'c', 'd'], 
        'col3': [1.1, 2.2, 3.3, np.nan]}
df = pd.DataFrame(data)

print("原始 DataFrame：")
print(df)
```

```python
# 刪除任何包含 NaN 的列
df_cleaned = df.dropna()  # 預設是 axis=0, how='any'
print("\n刪除包含 NaN 的列後的 DataFrame：")
print(df_cleaned)
```

```python
# 直接在原 DataFrame 上刪除包含 NaN 的列
df.dropna(inplace=True)
print("\n在原 DataFrame 上刪除包含 NaN 的列後的 DataFrame：")
print(df)
```

```python
# 只有當該列的所有值都是 NaN 時，才刪除該列
data2 = {'col1': [1, 2, np.nan, 4], 
        'col2': ['a', np.nan, np.nan, 'd'], 
        'col3': [1.1, 2.2, np.nan, np.nan]}
df2 = pd.DataFrame(data2)
print("\n原始 DataFrame2：")
print(df2)
```

```python
df2_cleaned = df2.dropna(how='all')
print("\n刪除所有值都是 NaN 的列後的 DataFrame2：")
print(df2_cleaned)
```

```python
# 指定要檢查的列名列表
data3 = {'col1': [1, 2, np.nan, 4], 
        'col2': ['a', np.nan, 'c', 'd'], 
        'col3': [1.1, 2.2, 3.3, np.nan]}
df3 = pd.DataFrame(data3)
print("\n原始 DataFrame3：")
print(df3)
```

```python
df3_cleaned = df3.dropna(subset=['col1','col3'])
print("\n刪除col1或col3包含NaN的列後的 DataFrame3：")
print(df3_cleaned)
```

```python
import pandas as pd

def main():
    """
    主函數，用於讀取CSV文件，進行數據清洗和轉換，並將結果保存為CSV和Excel文件。
    """
    # 讀取名為'上市公司資料.csv'的CSV文件，並將數據存儲在DataFrame對象df中。
    try:
        df = pd.read_csv('上市公司資料.csv')
    except FileNotFoundError:
        print("Error: 上市公司資料.csv not found.")
        return
    
    # 移除df中包含任何NaN（缺失值）的行，並將結果存儲在新的DataFrame對象df1中。
    df1 = df.dropna()
    
    # 重新索引df1的列，只保留指定的列，並將結果存儲在新的DataFrame對象df2中。
    # 保留的列包括：'公司代號'、'出表日期'、'公司名稱'、'產業別'、'營業收入-當月營收'、'營業收入-上月營收'。
    df2 = df1.reindex(columns=['公司代號','出表日期','公司名稱','產業別','營業收入-當月營收','營業收入-上月營收'])
    
    # 重命名df2中的列名，將'營業收入-上月營收'改為'上月營收'，將'營業收入-當月營收'改為'當月營收'，並將結果存儲在新的DataFrame對象df3中。
    df3 = df2.rename(columns={
        '營業收入-上月營收':'上月營收',
        '營業收入-當月營收':'當月營收'
        })
    
    # 將df3的數據保存到名為'上市公司資料整理.csv'的CSV文件中，並使用utf-8編碼。
    df3.to_csv('上市公司資料整理.csv',encoding='utf-8', index=False)
    
    # 將df3的數據保存到名為'上市公司資料整理.xlsx'的Excel文件中。
    # 修正: to_excel() 不支援 encoding 參數
    df3.to_excel('上市公司資料整理.xlsx', index=False)
    
    # 印出"存檔完成"的字串，表示檔案儲存完成。
    print("存檔完成")

# 檢查是否直接執行此腳本。如果是，則執行main()函數。
if __name__ == '__main__':
    main()
```

```python
import numpy as np
def str_to_int(value:str) -> int:
    try:
        value_int = int(value)
    except:
        value_int = 0
    return value_int

def str_to_float(value:str) -> float:
   try:
        value_float = float(value)
   except:
        value_float = 0
   return value_float

df3['年底人口數'] = df3['年底人口數'].map(str_to_int)
df3['人口密度'] = df3['人口密度'].map(str_to_int)
df3['土地面積'] = df3['土地面積'].map(str_to_float)
```

```python
def split(value:str) -> str:
    if len(value) > 4:
        return value[:3]
    else:
        return value
    
df3['縣市'] = df3['區域別'].map(split)
```

```python
df4 = df3.reindex(columns=['縣市','區域別','年底人口數','土地面積','人口密度'])
df4
```

```python
def population(value:int) -> str:
    if value > 367653:
        return '人多'
    elif value > 183826:
        return '適中'
    else:
        return '人少'
    
df4['人口分佈'] = df4['年底人口數'].map(population)
df4
```

```python
import numpy as np
import pandas as pd

scores = np.random.randint(50, 101, size=(50, 5))
df = pd.DataFrame(scores,
             columns=['國文', '英文', '數學', '地理', '歷史'],
             index=range(1,51)
             )
df.index.name = '學號'
df.columns.name = '科目'
df
```

```python
sum_value = df.sum(axis=1)
mean_value = df.mean(axis=1)
```

```python
df['總分'] = sum_value
df['平均'] = mean_value
```

```python
df1 = df.sort_values(by='總分',ascending=False)
df1
```

```python
rand_value = df1['總分'].rank(ascending=False,method='min')
df1['排名'] = rand_value
df1
```

```python
def max_score(row):
    less_60_count = len(row[row < 60])
    return pd.Series([row.min(),row.max(),row.std(),less_60_count],index=['最低分','最高分','標準差','不及格的數量'])
df2 = df1[['國文', '英文', '數學', '地理', '歷史']].apply(max_score,axis=1)
pd.concat([df1,df2],axis=1)
```
